Open an image from an URL

```python
# Complete the missing code to ensure that all the cells from Google Colab
# can be executed correctly.

# Complete the XXXXX to import the following libraries that will be needed for
# execute the code.
import urllib.request
from PIL import Image, ImageOps # From PIL, import Image and ImageOps
import matplotlib.pyplot as plt # import pyplot from matplotlib as plt
import numpy as np # import numpy library to use it only using "np"

# Code to read an image from an URL
urllib.request.urlretrieve('https://www.universalpictures.es/tl_files/content/movies/minions2/gallery/03.jpg', "minions.jpg")

# Use the function open from Image to open the image previously loaded with 
# the previous function.
img_color = Image.open("minions.jpg")

# Check that, after executing this block of code, the message Image read!!! is 
# printed below.
print("Image read!!!!")
```

Plot the original image and in grayscale

```python
# Use show function from pyplot(plt) to display the image you just read
plt.imshow(img_color)
plt.show()

# Convert the color image  to grayscale using the convert method from img_color variable
grayscale_image = img_color.convert("L")
grayscale_array = np.asarray(grayscale_image)
plt.imshow(grayscale_array, cmap="gray")
```

Generate the histogram of the previous image and display it

```python
# Plot the histogram of the img_color read using the histogram function from numpy (np)
histogram, bin_edges = np.histogram(img_color, bins=256, range=(0, 255))
plt.plot(bin_edges[0:-1], histogram) 

# Display the histogram using the show function from pyplot (plt)
plt.show()
```

Equalize the histogram and see the differences before/after equalization
